```python
from pdfminer.high_level import extract_text
text = extract_text('/content/HDFC_FQA.pdf')
print(text)
```

# Tokenization

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
tokenizer.word_index
```

```python
print(len(tokenizer.word_index)) # Total number of unique words i.e vocabulary
```

# Arranging the words into sequence

```python
input_sequences = []
for line in text.split('\n'):
  token_list = tokenizer.texts_to_sequences([line])[0]
  for i in range(1, len(token_list)):
    input_sequences.append(token_list[:i+1])
input_sequences
```

# Padding

```python
print(len(input_sequences[0]))
print(len(input_sequences[230]))
print(len(input_sequences[1100]))
print(len(input_sequences[670]))
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')
padded_input_sequences
```

```python
len(padded_input_sequences[0])
len(padded_input_sequences[100])
len(padded_input_sequences[1300])
len(padded_input_sequences[120])
len(padded_input_sequences[1100])
```

```python
x = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
x.shape, y.shape
```

```python
y
```

# Changing the y array into binary array i.e 0 and 1

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=1370)
```

```python
y
```

# Building LSTM model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
from logging import logProcesses
model = Sequential()
model.add(Embedding(1370, 100, input_length= max_len-1 ))
model.add(LSTM(128))
model.add(Dense(1370, activation = 'softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x,y,epochs = 100)
```

# Predicting the next word

```python
import time
import numpy as np
text = "What are the Key features of the HDFC Bank SmartWealth?"

for i in range(100):
  token_text = tokenizer.texts_to_sequences([text])[0]
  print(token_text)
  padded_token_text = pad_sequences([token_text], maxlen=max_len, padding='pre')
  predict_token= model.predict(padded_token_text)
  print("Predict_token: ", predict_token)
  predicted = np.argmax(predict_token)
  print("predicted: ",predicted)

  for word, index in tokenizer.word_index.items():
    # print("word:", word+" index: ", index)
    if index == predicted:
      text = text+" "+word
      print(text)
      time.sleep(2)
```
